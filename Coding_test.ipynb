## Coding test 1

Question:
    https://leetcode.com/problems/fizz-buzz/
    

````python
def fizz_buzz(n):
    """ This is a FizzBuzz Coding Test 
    :type n: int
    :return result type: list
    """
    
    result = []
    
    for i in range(1, n+1):
        
        if i % 3 == 0 and i % 5 == 0:
            result.append("FizzBuzz")
        elif i % 3 == 0:
            result.append("Fizz")
        elif i % 5 ==0:
            result.append("Buzz")
        else:
            result.append(str(i))
    
    return result
````

````python
fizz_buzz(5)
````
